```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('sample.csv')
data.shape
data
```

```python

data['lable'] = pd.Categorical(
    data['lable'],
    categories=[1,2,3,4]
)

df2 = pd.get_dummies(data['lable'], dtype = int)
```

```python

data = pd.concat([data,df2], axis=1)

data
```

```python
data = data.drop(['lable',], axis=1)
```

```python
x = data.iloc[0, :11]
y = data.iloc[0,11:]
x = np.array(x)
y = np.array(y)

print(x)
print('\n')
print(y)
```
